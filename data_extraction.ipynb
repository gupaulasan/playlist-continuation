```python
def normalize_name(name):
    name = name.lower()
    name = re.sub(r"[.,#!$%\^\*;:{}=\_`~()@]", ' ', name)
    name = re.sub(r'\s+', ' ', name).strip()
    return name
```

```python
subset_10_000 = pd.read_pickle(f"./data/raw/subset_10_000.pkl")
```

```python
subset_10_000["track_uri"] = subset_10_000["track_uri"].map(track_uri_mapping)
subset_10_000["artist_uri"] = subset_10_000["artist_uri"].map(artist_uri_mapping)
subset_10_000["plist_name"] = subset_10_000["plist_name"].apply(normalize_name)

subset_10_000["pid"] = pd.to_numeric(subset_10_000["pid"], downcast="integer")
subset_10_000["track_uri"] = pd.to_numeric(subset_10_000["track_uri"], downcast="integer")
subset_10_000["artist_uri"] = pd.to_numeric(subset_10_000["artist_uri"], downcast="integer")

subset_10_000.to_pickle(f"./data/processed/subset_10_000.pkl")
```
